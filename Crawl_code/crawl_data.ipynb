```python
import pandas as pd

# Lấy danh sách các công ty đang niêm yết
stocks_df = screener_df['ticker']

# In ra số lượng mã cổ phiếu
print(f"Tổng số mã cổ phiếu niêm yết: {len(stocks_df)}")
```

```python
# Lấy cột mã cổ phiếu (symbol)
symbols = stocks_df.tolist()

# Ghi toàn bộ danh sách mã cổ phiếu vào file text
with open("stock-symbols.txt", "w", encoding="utf-8") as f:
    for symbol in symbols:
      if symbol in ['ATP', 'BCH', 'CAF', 'CBV', 'CCS', 'DHQ', 'E1VFVN30',
                  'EGL', 'ESL', 'FHH', 'FUCTVGF3', 'FUCTVGF4', 'FUCTVGF5',
                  'FUCVREIT', 'FUEABVN', 'FUEABVND', 'FUEBFVND', 'FUEDCMID',
                  'FUEFCV50', 'FUEIP100', 'FUEKIV30', 'FUEKIVFS', 'FUEKIVND',
                  'FUEMAV30', 'FUEMAVND', 'FUESSV30', 'FUESSV50', 'FUESSVFL',
                  'FUETCC50', 'FUETPVND', 'FUEVFVND', 'FUEVN100', 'FVN']:
        continue
      f.write(symbol + "\n")
```

```python
from vnstock import Quote
import os
import time
```

```python
SYMBOL_FILE = '/content/drive/MyDrive/BigData/stock-symbols.txt'
OUTPUT_DIR = '/content/drive/MyDrive/BigData/data'
SLEEP_TIME = 1
```

```python
with open(SYMBOL_FILE, 'r', encoding='utf-8') as f:
        symbols = [line.strip().upper() for line in f if line.strip()]
print(len(symbols))
```

```python
for symbol in symbols:
    # if symbol in ['ATP', 'BCH', 'CAF', 'CBV', 'CCS', 'DHQ', 'E1VFVN30',
    #               'EGL', 'ESL', 'FHH', 'FUCTVGF3', 'FUCTVGF4', 'FUCTVGF5',
    #               'FUCVREIT', 'FUEABVN', 'FUEABVND', 'FUEBFVND', 'FUEDCMID',
    #               'FUEFCV50', 'FUEIP100', 'FUEKIV30', 'FUEKIVFS', 'FUEKIVND',
    #               'FUEMAV30', 'FUEMAVND', 'FUESSV30', 'FUESSV50', 'FUESSVFL',
    #               'FUETCC50', 'FUETPVND', 'FUEVFVND', 'FUEVN100', 'FVN']:
    #     continue
    output_file = os.path.join(OUTPUT_DIR, f"{symbol}.csv")
    if os.path.exists(output_file):
        continue
    quote = Quote(symbol=symbol, source='TCBS')
    data = quote.history(start='2020-01-01', end='2025-10-23')
    data.to_csv(output_file, index=False)
    print(symbol)
```
